```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import math
```

$ y = sin(x) $

```python
x = [i for i in range(-10, 10)]
y_sine = [math.sin(i) for i in x]
```

```python
plt.plot(x, y_sine)
```

$y = x^3$

```python
y_neg_cube = [-i ** 3 for i in x]
```

```python
plt.plot(x, y_neg_cube)
```

```python
# plt.figure(figsize=(15, 5))
plt.plot(x, y_neg_cube)
plt.title("Negative Cube")
plt.xlabel("X Values")
plt.ylabel("Negative cube of X")
```

```python
plt.plot(x, y_neg_cube)
y_sine_500 = [500* i for i in y_sine]
plt.plot(x, y_sine_500)
plt.legend(["Negative Cube", "500*Sine"])
```
