```python
import matplotlib.pyplot as plot
import numpy as np
%matplotlib inline
```

# Train Image

## Read

```python
train_img_1 = plot.imread('./Dataset/TestData/apple10-000-000.png')
train_img_2 = plot.imread('./Dataset/TestData/apple10-022-000.png')
```

```python
plot.subplot(1, 2, 1)
plot.imshow(train_img_1)
plot.subplot(1, 2, 2)
plot.imshow(train_img_2)
```

## Feature Extract

```python
train_img_gray_1 = np.mean(train_img_1, axis=2)
train_img_gray_2 = np.mean(train_img_2, axis=2)

assert train_img_gray_1.shape == train_img_gray_2.shape == (256, 256)
```

```python
plot.subplot(1, 2, 1)
plot.imshow(train_img_gray_1, cmap='gray')
plot.subplot(1, 2, 2)
plot.imshow(train_img_gray_2, cmap='gray')
```

```python
mean_img_1 = np.mean(train_img_gray_1)
mean_img_2 = np.mean(train_img_gray_2)
```

```python
print("Means are:", (mean_img_1, mean_img_2))
```

```python
std_img_1 = np.std(train_img_gray_1)
std_img_2 = np.std(train_img_gray_2)
```

```python
print("Standard Deviations are:", (std_img_1, std_img_2))
```

# Test Image

## Read

```python
test_img = plot.imread('./Dataset/TestData/apple10-022-090.png')
plot.imshow(test_img)
```

```python
test_img_gray = np.mean(test_img, axis=2)
assert test_img_gray.shape == (256, 256)
```

## Feature Extract

```python
mean_test = np.mean(test_img_gray)
std_test = np.std(test_img_gray)
```

```python
print("Test mean:", mean_test)
print("Test standard deviation:", std_test)
```

# Compare Train And Test

```python
cb_dist_1 = abs(mean_img_1 - mean_test) + abs(std_img_1 - std_test)
cb_dist_2 = abs(mean_img_2 - mean_test) + abs(std_img_2 - std_test)
```

```python
print("City block distance from img1:", cb_dist_1)
print("City block distance from img2:", cb_dist_2)
```

```python
if cb_dist_1 < cb_dist_2:
    print("Most similar to image 1")
elif cb_dist_2 < cb_dist_1:
    print("Most similar to image 2")
else:
    print("Likely similar to both the images")
```
