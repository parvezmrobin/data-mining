```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Input Image

```python
img = plt.imread('./Dataset/TrainData/apple1-000-000.png')
```

```python
plt.imshow(img)
```

## Separate Channels

```python
red = img[:, :, 0]
green = img[:, :, 1]
blue = img[:, :, 2]
```

### Check channels

```python
print("Red Channel")
plt.imshow(red, cmap='gray')
```

```python
print("Green Channel")
plt.imshow(green, cmap='gray')
```

```python
print("Blue Channel")
plt.imshow(blue, cmap='gray')
```

```python
red_vector = np.reshape(red, (-1))
green_vector = np.reshape(green, (-1))
blue_vector = np.reshape(blue, (-1))
assert red_vector.shape == green_vector.shape == blue_vector.shape
```

# Mean

```python
red_mean = np.mean(red)
green_mean = np.mean(green)
blue_mean = np.mean(blue)
print("Red mean:", red_mean)
print("Green mean:", green_mean)
print("Blue mean:", blue_mean)
```

# Median

```python
n = red_vector.shape[0]
print("Length of vector:", n)
```

```python
sorted_red = np.sort(red_vector)
sorted_green = np.sort(green_vector)
sorted_blue = np.sort(blue_vector)

n2 = int(n / 2)

median = lambda v: (v[n2] + v[n2 - 1]) / 2
red_median = median(sorted_red)
green_median = median(sorted_green)
blue_median = median(sorted_blue)

print("Red Median:", red_median)
print("Green Median:", green_median)
print("Blue median:", blue_median)
```

# Midrange

```python
mid_range = lambda v: (v[0] + v[n-1]) / 2

red_mid_range = mid_range(sorted_red)
green_mid_range = mid_range(sorted_green)
blue_mid_range = mid_range(sorted_blue)

print("Red mid range:", red_mid_range)
print("Green mid range:", green_mid_range)
print("Blue mid range:", blue_mid_range)
```

# Min

```python
red_min = sorted_red[0]
green_min = sorted_green[0]
blue_min = sorted_blue[0]

print("Red min:", red_min)
print("Green min:", green_min)
print("Blue min:", blue_min)
```

# Max

```python
red_max = sorted_red[n-1]
green_max = sorted_green[n-1]
blue_max = sorted_blue[n-1]

print("Red max:", red_max)
print("Green max:", green_max)
print("Blue max:", blue_max)
```

# Range

```python
red_range = (red_min, red_max)
green_range = (green_min, green_max)
blue_range = (blue_min, blue_max)

print("Red range:", red_range, "-", red_max - red_min)
print("Green range:", green_range, "-", green_max - green_min)
print("Blue range:", blue_range, "-", blue_max - blue_min)
```

# Standard Deviation

```python
def std_dev(v):
    mean = np.mean(v)
    squared = v * v
    variance = (np.sum(squared) / n) - (mean * mean)
    return np.sqrt(variance)
```

```python
red_std_dev = std_dev(red_vector)
green_std_dev = std_dev(green_vector)
blue_std_dev = std_dev(blue_vector)

print("Standard Deviation of Red:", red_std_dev)
print("Standard Deviation of Green:", green_std_dev)
print("Standard Deviation of Blue:", blue_std_dev)
```
